```python
import openai
openai.api_key = "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"
import csv
model_engine = "text-davinci-003"

# import openai

# # Set OpenAI API key and model engine
# openai.api_key = "YOUR_API_KEY_HERE"
# model_engine = "text-davinci-002"

# Read data from input CSV file
data = []
with open('Blue - URLs for Meta Content Testing.csv', 'r', newline='', encoding='utf-8') as file:
    reader = csv.reader(file)
    next(reader) # Skip header row
    for row in reader:
        address = row[0]
        meta_description = row[1]
        row_data = {'Address': address, 'Meta Description': meta_description}
        data.append(row_data)

# Generate prompts
prompts = []
for row_data in data:
    prompt = "The URL is: " + row_data['Address'] + " and the existing meta description is: " + row_data['Meta Description'] + ". Can you write a better meta description in Chinese with 80 Chinese characters?"
    prompts.append(prompt)

# Call OpenAI API and store response data in dictionary
responses = {}
for i, prompt in enumerate(prompts):
    completion = openai.Completion.create(
        engine=model_engine,
        prompt=prompt,
        max_tokens=1024,
        n=1,
        stop=None,
        temperature=0.5
    )
    response_text = completion.choices[0].text.strip()
    responses[data[i]['Address']] = response_text
    if(len(responses[data[i]['Address']])>80):
        completion = openai.Completion.create(
        engine=model_engine,
        prompt="Make it more natural and shorter and more precise, based on"+responses[data[i]['Address']]+"and make sure it is in 中文",
        max_tokens=1024,
        n=1,
        stop=None,
        temperature=0.5
    )
        response_text = completion.choices[0].text.strip()
        responses[data[i]['Address']] = response_text
    print(responses[data[i]['Address']])

# Write response data to CSV file
with open('Blue - output.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Address', 'Response'])
    for address, response in responses.items():
        writer.writerow([address, response])

```
